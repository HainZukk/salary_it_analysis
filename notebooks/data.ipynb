```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/salaries.csv")
```

Most salary in USD

```python
df.sort_values("salary_in_usd",ascending=False).iloc[0]
```

Min salary in USD

```python
df.sort_values("salary_in_usd",ascending=True).iloc[0]
```

Recuento total de personas por titulo

```python
df.value_counts("job_title").head(10)
```

Recuento de personas por pais de residencia

```python
df.value_counts("employee_residence").head(10)
```

Recuento de salarios argentinos

```python
ARGENTINA = (df["employee_residence"] == "AR").sum()
print(f"Cantidad de salarios en Argentina {ARGENTINA}")
```

Salarios mayores a $100.000 USD (En este Dataset)

```python
sueldo_mayor_argentina = ((df["employee_residence"] == "AR") & (df["salary_in_usd"] > 100000)).sum()
print(f"En argentina existen tan solo {sueldo_mayor_argentina} personas que tienen un sueldo mayor a $100.000 dolares anuales")
```

Recuento de Argentinos por nivel de experiencia

```python
df[df["employee_residence"] == "AR"]["experience_level"].value_counts()
```

Promedio de salario de Estados Unidos vs Argentina (En Dolares)

```python
df.groupby("employee_residence")["salary_in_usd"].mean().loc[["US","AR"]]
```

Mediana de salario de Estados Unidos vs Argentina (En dolares)

```python
df.groupby("employee_residence")["salary_in_usd"].median().loc[["US","AR"]]
```
