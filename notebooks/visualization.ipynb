```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/salaries.csv")
```

```python
sns.set_theme(style ="whitegrid")
```

Mediana de Salarios de Estados Unidos y Argentina

```python
sns.barplot(
    data=df[df["employee_residence"].isin(["US", "AR"])],
    x="employee_residence",
    y="salary_in_usd",
    estimator= "median",
    palette="viridis",
    hue="employee_residence",
    legend=False
)
plt.title("Mediana US vs AR")
plt.ylabel("Salario")
plt.xlabel("Pais de Residencia")
```

Promedio de Salarios de Estados Unidos y Argentina

```python
sns.barplot(
    data=df[df["employee_residence"].isin(["US", "AR"])],
    y="salary_in_usd",
    x="employee_residence",
    hue = "employee_residence",
    palette="viridis",
    legend=False
)

plt.title("Promedio US vs AR")
plt.ylabel("Salario Promedio")
plt.xlabel("Pais de Residencia")
```

Top 10 Roles en Argentina

```python
ar_titles = df[df["employee_residence"] == "AR"]
plt.figure(figsize=(10, 6))

sns.countplot(
    data=ar_titles,
    y="job_title",
    order=ar_titles["job_title"].value_counts().head(10).index,
    hue="job_title",
    palette="viridis",
    legend=False
)

plt.title("Top 10 roles en Argentina")
plt.ylabel("Puesto")
plt.xlabel("Personas")
plt.tight_layout()
```

Top 10 Roles en Estados Unidos

```python
ar_titles = df[df["employee_residence"] == "US"]
plt.figure(figsize=(10, 6))

sns.countplot(
    data=ar_titles,
    y="job_title",
    order=ar_titles["job_title"].value_counts().head(10).index,
    hue="job_title",
    palette="Set2",
    legend=False
)

plt.title("Top 10 roles en Estados Unidos")
plt.ylabel("Puesto")
plt.xlabel("Personas")
plt.tight_layout()
```

Salario por pais

```python
plt.Figure(figsize=(10, 6))
sns.boxplot(
    data=df[df["employee_residence"].isin(["US", "AR", "CA", "DE"])],
    x="employee_residence",
    y="salary_in_usd",
    hue="employee_residence",
    legend=False,
    palette="Set2"
)

plt.title("Salario por pais")
plt.ylabel("Salario en USD")
plt.xlabel("Pais de residencia")
```

Salario por Experiencia

```python
sns.barplot(
    data = df,
    x="experience_level",
    y = "salary_in_usd",
    estimator="median",
    palette="viridis",
    hue="experience_level",
    legend=False
)
```

Remote Ratio

```python
df_remote = (
    df.groupby("remote_ratio")["salary_in_usd"]
    .median()
    .reset_index()
)

sns.barplot(
    data=df_remote,
    y="remote_ratio",
    x="salary_in_usd",
    hue="remote_ratio",
    palette="viridis",
    legend=False
)

plt.title("Salario mediano segn modalidad de trabajo")
plt.xlabel("Salario en USD")
plt.ylabel("Nivel de trabajo en remoto")
plt.tight_layout()
```
